```python
tm = 1                                  # Tempo de vida médio
t0 = 0                                  # Tempo inicial 
tf = 5                                  # Tempo final de análise
dt = [0.01,0.02,0.04,0.06,0.08,0.1]     # Delta t : intervalo de variação do tempo 
N_0 = 100                               # Quant. inicial de elementos no tempo t0

N = []
temp = []
mat_delta = []
DN = []
DNp = []
DNr = []
DNc = []
mat_int_err_d1yp = [] 
mat_int_err_d1yr = []
mat_int_err_d1yc = []
mat_err_d1yp = []
mat_err_d1yr = []
mat_err_d1yc = [] 


for j in range(len(dt)):
    t = np.arange(t0,tf,dt[j])
    y  = N_0*np.exp(-t/tm)
    d1y = -y/tm
    
    n = len(t)
    
    d1yp = [(y[i+1]-y[i])/(t[i+1]-t[i]) for i in range(0,n-1)]

    d1yr = [(y[i]-y[i-1])/(t[i]-t[i-1]) for i in range(1,n)]

    d1yc = [(y[i+1]-y[i-1])/(t[i+1]-t[i-1]) for i in range(1,n-1)]

    err_d1yp = np.abs(d1yp-d1y[0:(n-1)]) 
    err_d1yr = np.abs(d1yr-d1y[1:n])
    err_d1yc = np.abs(d1yc-d1y[1:(n-1)])

    delta = (t[1] - t[0])
    int_err_d1yp = np.sum(err_d1yp)
    int_err_d1yr = np.sum(err_d1yr)
    int_err_d1yc = np.sum(err_d1yc)
    
    N.append(n)
    temp.append(t)
    mat_delta.append(delta)
    DN.append(d1y)
    DNp.append(d1yp)
    DNr.append(d1yr)
    DNc.append(d1yc)
    mat_err_d1yp.append(err_d1yp)
    mat_err_d1yr.append(err_d1yr)
    mat_err_d1yc.append(err_d1yc)
    mat_int_err_d1yp.append(int_err_d1yp)
    mat_int_err_d1yr.append(int_err_d1yr)
    mat_int_err_d1yc.append(int_err_d1yc)
```

```python
tm = 1                                  # Tempo de vida médio
t0 = 0                                  # Tempo inicial 
tf = 5                                  # Tempo final de análise
dt = [0.01,0.02,0.04,0.06,0.08,0.1]     # Delta t : intervalo de variação do tempo 
N_0 = 100                               # Quant. inicial de elementos no tempo t0

N_exat = []
temp = []
N_aprox = []
N_prog = []
N_central = []
N = []

for k in range(len(dt)):
    t = np.arange(t0,tf,dt[k])        
    y = N_0*np.exp(-t/tm)
    n = len(t)
    N_exat.append(y)
    N.append(n)
    temp.append(t)
# ------------------------------------------------------------
    y_aprox_a = [N_0]
    for a in range(len(t)-1):
        Nva = y_aprox_a[a] - (y_aprox_a[a]/tm)*(dt[k])
        y_aprox_a.append(Nva)
    
    erroa = [0]
    for j in range(len(y)-1):
        ea = np.abs(y[j]-y_aprox_a[j])
        erroa.append(ea)
    
    N_aprox.append([y_aprox_a,erroa])
    
# --------------------------------------------------------------
    y_aprox_p = [N_0]
    for b in range(len(t)-1):
        Nvb = y_aprox_p[b]*(1 - (dt[k]/tm))
        y_aprox_p.append(Nvb)
        
    errob = [0]
    for j in range(len(y)-1):
        eb = np.abs(y[j]-y_aprox_p[j])
        errob.append(eb)
    
    N_prog.append([y_aprox_p,errob])

# --------------------------------------------------------------
    y_aprox_c = [N_0,y[1]]
    for c in range(1,len(t)-2):
        Nvc = y_aprox_c[c-1] - (2*dt[k]/tm)*y_aprox_c[c]
        y_aprox_c.append(Nvc)
        
    erroc = [0]
    for j in range(len(y)-1):
        ec = np.abs(y[j]-y_aprox_c[j])
        erroc.append(ec)
    
    N_central.append([y_aprox_c,erroc])
```

O método de derivação progressiva apresenta os mesmos resultados do modelo de Euler, quando aplica-se a derivada disponibilizada do problema proposta, na equação resultante do truncamento da expansão de Taylor na segunda ordem.

### Questão 1.4

O sistema analisado comporta-se segundo o esquema mostrado abaixo:  

$N_A  \stackrel{k_1}\longrightarrow  N_B$

$N_B \stackrel{k_2}\longrightarrow  \emptyset$ 

Deste modo a matriz dos coeficientes estequiométrica (S) resultará em (1);

$ S = 
\left[
\begin{array}{ccc}
-1 & 0 \\
1 & -1 
\end{array}
\right]
 \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  (1) $  
 
onde as colunas representam as reações e as linhas, o comportamento dos elementos quando tratados os seus saldos ao final da respectiva reação. Uma segunda matriz é montada com o objetivo de descrever as quantidades de cada elemento como também a "velocidade" com que cada reação ocorre, a matriz $v$ (2).

$ v = 
\left[
\begin{array}{ccc}
k_1N_A  \\
k_2N_B 
\end{array}
\right]
 \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  (2)$ 

onde cada linha representa uma reação e cada elemento desta é oriunto da multiplicação da "constante de ocorrência" da reação pela concentração de cada um dos elementos participantes da reação pelo lado dos reagentes. Caso um dos coeficientes dos elementos analisados seja diferente de um, deste será elevado a potência do valor correspondente, mantendo a multiplicação com os demais elementos reagentes. 

A taxa de variação descrita na questão 1.4, é obtida realizando o produto matricial entre $S \cdot v$, resultando em uma nova matriz, com apenas uma coluna e com a quantidade de linha semelhante ao número de elementos analisados, o que podemos também chamar de vetor coluna, que atribuiremos dN (3), logo. 

$ dN = 
\left[
\begin{array}{ccc}
-1 & 0 \\
1 & -1 
\end{array}
\right] \cdot
\left[
\begin{array}{ccc}
k_1N_A  \\
k_2N_B 
\end{array}
\right]
 \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  (3) $ 

que resultará em (4)

$ dN = 
\left[
\begin{array}{ccc}
-k_1N_A  \\
k_1N_A\ -\ k_2N_B 
\end{array}
\right]
 \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  (4) $ 

onde 

$\dfrac{d N_A}{dt} =\ -k_1N_A \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  (5) $

$\dfrac{d N_B}{dt} = \ k_1N_A - \ K_2N_B \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  (6)$ 

considerando $k_1\ =\dfrac{1}{\tau_A}$ e $k_2\ =\dfrac{1}{\tau_B}$, obtemos (7) e (8)

$\dfrac{d N_A}{dt} = \dfrac{-N_A}{\tau_A} \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \    (7)$

$\dfrac{d N_B}{dt} = \dfrac{N_A}{\tau_A} - \dfrac{N_B}{\tau_B} \ \ \ \ \ \ \ \ \ \     (8)$       

Resolvendo as equações (7) e (8) segundo os métodos adequados, obtendo (9) e (10) respectivamente

$N_A(t) = N_Ae^{-t/ta} \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ (9)$ 

$N_B(t) = N_Be^{-t/tb} + \dfrac{1}{\dfrac{ta}{tb} - 1}(e^{-t/ta} - e^{-t/tb})N_A \ \ \ \ \ (10)$

Entretanto para descrever o modelo recorreremos a algoritmos, que por meio do método de Euler, resultante da expansão de Taylor para ambas as funções: $N_A(t)\ e\ N_B(t)$ descreveram o crescimento ou descrescimento das "populações" analisadas. Deste modo pelo método de Euler tem-se (11):

$N(t+\Delta t)\ =\ N(t)\ + \dfrac{dN(t)}{dt}\ \Delta\ t  \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  (11)$ 

Substituindo (7) e (8) em (11), obtem-se respectivamente (12) e (13)

$N_A(t+\Delta t)\ =\ N_A(t)\ - \dfrac{N_A}{\tau_A}\ \Delta\ t  \ \ \ \ \ \ \ \ \ \ \  \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  (12)$ 

$N_B(t+\Delta t)\ =\ N_B(t)\ + (\dfrac{N_A}{\tau_A} - \dfrac{N_B}{\tau_B})\ \Delta\ t  \ \ \ \ \ \ \ \ \ \ \ \ \ (13)$

```python
def eq1(tma,NA):                   # Taxa de variação para N_A
    eq = -NA/tma
    return eq

def eq2(tma,tmb,NA,NB):            # Taxa de variação para N_B
    eq = NA/tma - NB/tmb
    return eq

def metEuler(t0,tf,dt,Na,Nb,tma,tmb):
    t = np.arange(t0,tf,dt)
    n = len(t)
    N_a = [Na]
    N_b = [Nb]
    
    for i in range(0,n-1):
        k1 = eq1(tma,N_a[i])
        k2 = eq2(tma,tmb,N_a[i],N_b[i])
        NA = N_a[i] + k1*dt                  # Método de Euler para N_A
        NB = N_b[i] + k2*dt                  # Método de Euler para N_B
        N_a.append(NA)
        N_b.append(NB)
        
    return N_a,N_b 
```

```python
t0 = 0
tf = 1000
dt = 0.01
Na = 1000
Nb = 1000
ta = 10
tb = 100
h = ta/tb

t = np.arange(t0,tf,dt)

q = np.exp(-t/ta)
p = np.exp(-t/tb)
s = q - p

d = (1/(ta/tb - 1))*Na*s
e = Nb*np.exp(-t/tb)
r = e+d

N_A = Na*np.exp(-t/ta)
N_B = Nb*np.exp(-t/tb)+(1/((ta/tb)-1))*Na*(np.exp(-t/ta)-np.exp(-t/tb))


plt.figure(figsize=(20,7))

plt.subplot(1,3,1)
plt.plot(t,q, '-b',label = "ta")
plt.plot(t,p, '-r',label = "tb")
plt.plot(t,s, '-g',label = "ta - tb")
plt.legend(prop={'size':12})
plt.grid()

plt.subplot(1,3,2)
plt.plot(t,d, '-c',label = "dir -> NA")
plt.plot(t,e, '-m',label = "esq -> NB")
plt.plot(t,r, '-y',label = "NB")
plt.legend(prop={'size':12})
plt.grid()

plt.subplot(1,3,3)
plt.plot(t,N_A, '-b',label = "Na")
plt.plot(t,N_B, '-r',label = "Nb")
plt.title("ta/tb = %.2f" %h)
plt.xlabel('$Tempo$',fontsize = 14)
plt.ylabel('$Número\ de\ núcleos$',fontsize = 14)
plt.tick_params(axis = 'x',labelsize=12)
plt.tick_params(axis = 'y',labelsize=12)
plt.legend(prop={'size':12})
plt.grid()

plt.show()
```

### Questão 1.5

O sistema analisado comporta-se segundo o esquema mostrado abaixo:  

$N_A  \stackrel{k_1}\longrightarrow  N_B$

$N_B \stackrel{k_2}\longrightarrow  N_A$ 

Deste modo a matriz dos coeficientes estequiométrica (S) resultará em (1);

$ S = 
\left[
\begin{array}{ccc}
-1 & 1 \\
1 & -1 
\end{array}
\right]
 \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  (1) $  
 
onde as colunas representam as reações e as linhas, o comportamento dos elementos quando tratados os seus saldos ao final da respectiva reação. Uma segunda matriz é montada com o objetivo de descrever as quantidades de cada elemento como também a "velocidade" com que cada reação ocorre, a matriz $v$ (2).

$ v = 
\left[
\begin{array}{ccc}
k_1N_A  \\
k_2N_B 
\end{array}
\right]
 \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  (2)$ 

onde cada linha representa uma reação e cada elemento desta é oriunto da multiplicação da "constante de realização" da reação pela concentração de cada um dos elementos participantes da reação pelo lado dos reagentes. Caso um dos coeficientes dos elementos analisados seja diferente de um, deste será elevado a potência do valor correspondente, mantendo a multiplicação com os demais elementos reagentes. 

A taxa de variação descrita na questão 1.4, é obtida realizando o produto matricial entre $S \cdot v$, resultando em uma nova matriz, com apenas uma coluna e com a quantidade de linha semelhante ao número de elementos analisados, o que podemos também chamar de vetor coluna, que atribuiremos dN (3), logo. 

$ dN = 
\left[
\begin{array}{ccc}
-1 & 1 \\
1 & -1 
\end{array}
\right] \cdot
\left[
\begin{array}{ccc}
k_1N_A  \\
k_2N_B 
\end{array}
\right]
 \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  (3) $ 

que resultará em (4)

$ dN = 
\left[
\begin{array}{ccc}
k_2N_B\ -\ k_1N_A  \\
k_1N_A\ -\ k_2N_B 
\end{array}
\right]
 \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  (4) $ 

onde 

$\dfrac{d N_A}{dt} = \ k_2N_B - \ k_1N_A \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  (5) $

$\dfrac{d N_B}{dt} = \ k_1N_A - \ k_2N_B \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  (6)$ 

considerando $k_1\ =\dfrac{1}{\tau_A}$ e $k_2\ =\dfrac{1}{\tau_B}$, obtemos (7) e (8)

$\dfrac{d N_A}{dt} = \dfrac{N_B}{\tau_B} - \dfrac{N_A}{\tau_A} \ \ \ \ \ \ \ \ \ \    (7)$

$\dfrac{d N_B}{dt} = \dfrac{N_A}{\tau_A} - \dfrac{N_B}{\tau_B} \ \ \ \ \ \ \ \ \ \     (8)$       

Para resolvermos as equações (7) e (8) segundo os métodos adequados, resutando em 

$N_A(t) = \dfrac{N_A\ -\ RN_B}{R\ +\ 1}\ (1 \ +\ e^{-(R+1)\dfrac{t}{\tau_A}}) + N_B \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ (9)$ 

$N_B(t) = \dfrac{N_A\ -\ RN_B}{R\ +\ 1}\ (R \ -\ e^{-(R+1)\dfrac{t}{\tau_A}}) + RN_B  \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ (10)$

onde R = $\dfrac{\tau_A}{\tau_B}$. Para simplificar no modelo admitiremos o comportamento para R = 1, ou seja, $\tau_A$ = 
$\tau_B$, nessa circuntância obteremos as equações (11) e (12) 

$N_A(t) = \dfrac{N_A\ -\ N_B}{2}(1 \ +\ e^{\dfrac{-2t}{\tau_A}}) + N_B \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ (12)$ 

$N_B(t) = \dfrac{N_A\ -\ N_B}{2}(1 \ -\ e^{\dfrac{-2t}{\tau_A}}) + N_B  \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ (12)$

Entretanto para descrever o modelo recorreremos a algoritmos, que por meio do método de Euler, resultante da expansão de Taylor para ambas as funções: $N_A(t)\ e\ N_B(t)$ descreveram o crescimento ou descrescimento das "populações" analisadas. Deste modo pelo método de Euler tem-se (11):

$N(t+\Delta t)\ =\ N(t)\ + \dfrac{dN(t)}{dt}\ \Delta\ t  \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  (13)$ 

Substituindo (7) e (8) em (11) tem-se (12) e (13), respectivamente

$N_A(t+\Delta t)\ =\ N_A(t)\ + (\dfrac{N_B}{\tau_B} - \dfrac{N_A}{\tau_A}) \ \Delta\ t  \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  (14)$ 

$N_B(t+\Delta t)\ =\ N_B(t)\ + (\dfrac{N_A}{\tau_A} - \dfrac{N_B}{\tau_B})\ \Delta\ t  \ \ \ \ \ \ \ \ \ \ \ \ \ (15)$

```python
def equac1(tm,NA,NB):
    eq = NB/tm - NA/tm
    return eq

def equac2(tm,NA,NB):
    eq = NA/tm - NB/tm
    return eq

def met_Euler(t0,tf,dt,Na,Nb,tm):
    t = np.arange(t0,tf,dt)
    n = len(t)
    N_a = [Na]
    N_b = [Nb]
    
    for i in range(0,n-1):
        k1 = equac1(tm,N_a[i],N_b[i])
        k2 = equac2(tm,N_a[i],N_b[i])
        NA = N_a[i] + k1*dt
        NB = N_b[i] + k2*dt
        N_a.append(NA)
        N_b.append(NB)
        
    return N_a,N_b 
```

```python
Na = 100
Nb = 0
tm = [0.5,1,2,5,10,20]
dt = 0.01
t0 = 0
tf = 20

t = np.arange(t0,tf,dt)
A = []
B = []

for j in range(len(tm)):
    a,b = met_Euler(t0,tf,dt,Na,Nb,tm[j])
    A.append(a)
    B.append(b)
    
plt.figure(figsize=(15,8))
   
plt.plot(t,A[0], '-b',label = r'$\tau = 0.5$')
plt.plot(t,B[0], '-b')

plt.plot(t,A[1], '-g',label =  r'$\tau = 1$')
plt.plot(t,B[1], '-g')

plt.plot(t,A[2], '-c',label = r'$\tau = 2$')
plt.plot(t,B[2], '-c')

plt.plot(t,A[3], '-r',label = r'$\tau = 5$')
plt.plot(t,B[3], '-r')

plt.plot(t,A[4], '-m',label = r'$\tau = 10$')
plt.plot(t,B[4], '-m')

plt.plot(t,A[5], '-y',label = r'$\tau = 20$')
plt.plot(t,B[5], '-y')

plt.xlabel('$Tempo$',fontsize = 14)
plt.ylabel('$Número\ de\ elementos$',fontsize = 14)
plt.tick_params(axis = 'x',labelsize=12)
plt.tick_params(axis = 'y',labelsize=12)
plt.title("Dados obtidos pelo método de Euler")
plt.legend(prop={'size':14})
plt.axis([t0,tf,0,100])
plt.grid()

plt.show()
```

### Comparação entre os dois modelos (equação analítica e o método de Euler) [ANÁLISE DOS ERROS]

```python
Na = 1000                                    # Valor inicial de N_A
Nb = 0                                       # Valor inicial de N_B
tm = [0.5,1,2,5,10,20] 
dt = 0.1
t0 = 0                                       # Tempo inicial
tf = 1000                                    # Tempo final

t = np.arange(t0,tf,dt)
A_eul_tm = []
B_eul_tm = []
A_an_tm = []
B_an_tm = []

for j in range(len(tm)):
    a,b = met_Euler(t0,tf,dt,Na,Nb,tm[j])
    A_eul_tm.append(a)
    B_eul_tm.append(b)
    
for k in range(len(tm)):
        u = t/tm[k]
        c = ((Na - Nb)/2)*(1 + np.exp(-2*u)) + Nb
        d = ((Na - Nb)/2)*(1 - np.exp(-2*u)) + Nb
        A_an_tm.append(c)
        B_an_tm.append(d)

Erro_A_tm = []
Erro_B_tm = []

for i in range(len(tm)):
        err_a_tm = []
        err_b_tm = []
        for h in range (len(t)):
            er_a_tm = A_an_tm[i][h] - A_eul_tm[i][h]
            er_b_tm = B_an_tm[i][h] - B_eul_tm[i][h]
            err_a_tm.append(er_a_tm)
            err_b_tm.append(er_b_tm)
            
        Erro_A_tm.append(err_a_tm)
        Erro_B_tm.append(err_b_tm)
```

### Funcão para plotar gráfico de f(x)

```python
def plotar(x_min,x_max,steps,f_x):
    f = np.zeros(steps)
    j = np.linspace(x_min,x_max,steps)
    
    for i in enumerate(j):
        f[i[0]] = f_x.subs(x,i[1])
        
    plt.plot(j,f, 'o-b')
    plt.grid()
    plt.show()
```

### Função para determinar a derivada de uma função $f(x)$ com respeito a $x$ e plotar o grafico de $\dfrac{dx}{dt}$

```python
def derivad(x_min,x_max,steps,f_x):
    x = sp.symbols('x')
    dfdx = sp.diff(f_x,x)
    j = np.linspace(x_min,x_max,steps)
    Df_array = np.zeros(steps)
    for i in enumerate(j):
        Df_array[i[0]] = dfdx.subs(x,i[1])

    return Df_array
```

```python
def funcao(x_min,x_max,steps,f_x):
    j = np.linspace(x_min,x_max,steps)

    x = sp.symbols('x')
    dfdx = sp.diff(f_x,x)

    Df_array = np.zeros(steps)
    fx = np.zeros(steps)

    for i in enumerate(j):
        Df_array[i[0]] = dfdx.subs(x,i[1])    
    
    for k in enumerate(j):
        fx[k[0]] = f_x.subs(x,k[1])    

    plt.figure(figsize=(12,5))

    plt.subplot(1,2,1)
    plt.plot(j,fx,'-b')
    plt.grid()
    plt.title('Função de f(x)')

    plt.subplot(1,2,2)
    plt.plot(j,Df_array,'-r')
    plt.grid()
    plt.title('Derivada de f(x)')

    plt.show()
```

```python
x_min =-5
x_max = 5
steps = 20
j = np.linspace(x_min,x_max,steps)
f_x = x**3 + x**2 -10*x + 10

x = sp.symbols('x')
dfdx = sp.diff(f_x,x)

Df_array = np.zeros(steps)
fx = np.zeros(steps)

for i in enumerate(j):
    Df_array[i[0]] = dfdx.subs(x,i[1])    
    
for k in enumerate(j):
    fx[k[0]] = f_x.subs(x,k[1]) 
    
d1yp = [(fx[i+1]-fx[i])/(j[i+1]-j[i]) for i in range(0,steps-1)]

d1yc = [(fx[i+1]-fx[i-1])/(j[i+1]-j[i-1]) for i in range(1,steps-1)]

plt.figure(figsize=(7,5))

plt.plot(j,Df_array,'-r',label = 'Der. analítica')
plt.plot(j[0:(steps-1)],d1yp,'o-b',label = 'Der. Num. Progressiva')
plt.plot(j[1:(steps-1)],d1yc,'o-g',label = 'Der. Num. Ponto Central')
plt.legend()
plt.grid()
plt.show()
```

```python
x_min =-5
x_max = 5
steps = 20
j = np.linspace(x_min,x_max,steps)
f_x = x**3 + x**2 -10*x + 10

x = sp.symbols('x')
dfdx = sp.diff(f_x,x)
df2dx = sp.diff(f_x,x,x)

Df_array = np.zeros(steps)
fx = np.zeros(steps)

for i in enumerate(j):
    Df_array[i[0]] = dfdx.subs(x,i[1])    
    
for k in enumerate(j):
    fx[k[0]] = f_x.subs(x,k[1]) 
    
delta = j[1]-j[0]
    
d1yp = [(fx[i+1]-fx[i])/(j[i+1]-j[i])-(df2dx.subs(x,j[i])/(2*delta))*(delta**2) for i in range(0,steps-1)]

d1yc = [(fx[i+1]-fx[i-1])/(j[i+1]-j[i-1]) for i in range(1,steps-1)]

plt.figure(figsize=(7,5))

plt.plot(j,Df_array,'-r',label = 'Der. analítica')
plt.plot(j[0:(steps-1)],d1yp,'o-b',label = 'Der. Num. Progressiva')
plt.plot(j[1:(steps-1)],d1yc,'o-g',label = 'Der. Num. Ponto Central')
plt.legend()
plt.grid()
plt.show()
```

## Questão 3 : Erro integrado

## a)  $y = x^3 + x^2 - 10 x + 10, ~x \in [-5,5]$

```python
x_min =-5
x_max = 5

mat_delta = []
mat_int_err_d1yp = [] 
mat_int_err_d1yc = []
mat_err_d1yp = []
mat_err_d1yc = [] 
mat_m = []
hipot1 = []
hipot2 = []
mat_err_d1yp_hip = []
mat_err_d1yc_hip = []

for j in range(10,120,20):
    
    m = np.linspace(x_min,x_max,j)
    f_x = x**3 + x**2 -10*x + 10

    x = sp.symbols('x')
    dfdx = sp.diff(f_x,x)

    Df_array = np.zeros(j)
    fx = np.zeros(j)

    for i in enumerate(m):
        Df_array[i[0]] = dfdx.subs(x,i[1])
    
    for k in enumerate(m):
        fx[k[0]] = f_x.subs(x,k[1]) 
    
    d1yp = [(fx[i+1]-fx[i])/(m[i+1]-m[i]) for i in range(0,j-1)]

    d1yc = [(fx[i+1]-fx[i-1])/(m[i+1]-m[i-1]) for i in range(1,j-1)]
    
    err_d1yp = np.abs(d1yp-Df_array[0:(j-1)]) 
    err_d1yc = np.abs(d1yc-Df_array[1:(j-1)])
    
    err_d1yp_hip = d1yp-Df_array[0:(j-1)]
    err_d1yc_hip = d1yc-Df_array[1:(j-1)]

    delta = m[1] - m[0]
    int_err_d1yp = np.sum(err_d1yp)
    int_err_d1yc = np.sum(err_d1yc)
    
    mat_delta.append(delta)
    mat_int_err_d1yp.append(int_err_d1yp)
    mat_int_err_d1yc.append(int_err_d1yc)
    mat_err_d1yp.append(err_d1yp)
    mat_err_d1yc.append(err_d1yc)
    mat_err_d1yp_hip.append(err_d1yp_hip)
    mat_err_d1yc_hip.append(err_d1yc_hip)
    mat_m.append(m)
    
for a in range(len(mat_delta)):
    hipot1.append(mat_delta[a]**2)
    
for b in range (len(mat_delta)):   
    hipot2.append(mat_delta[b]**3)
```

```python
# Carregando bibliotecas numérica, gráfica e simbólica

import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import numpy as np
import sympy as sp
sp.init_printing()

# Carrega a biblioteca de widgets
from ipywidgets import interact
import ipywidgets as widgets
```

```python
def integral(x,y,tipo):

    N = len(x)
    dx = x[1] - x[0]
    
    if tipo==1:
        tmp = 0
        for i in range(0,N):
            tmp = tmp + y[i]*dx

    elif tipo==2:
        tmp = 0
        for i in range(0,N):
            tmp = tmp + y[i]*dx
        
        tmp = tmp - (y[0]+y[N-1])*dx/2
    
    else:
        tmp = 0
        for i in range(0,N-1):
            tmp = tmp + (y[i]+y[i+1])*dx/2
            
    return tmp
```

## a)  $y = x^3 + x^2 - 10 x + 10, ~x \in [-5,5]$

```python
x1 = -5
x2 = 5
f_x = x**3 + x**2 - 10*x + 10

x = sp.symbols('x')
int_f_x = sp.integrate(f_x,(x,x1,x2))

mat_N = []
mat_dx = []
mat_met1 = []
mat_met2 = []
mat_met3 = []
mat_met1_e = []
mat_met2_e = []
mat_met3_e = []

for n in range(10,100,1):
    for j in [1,2,3]:
        
        m = np.linspace(x1,x2,n)
        dx = m[1] - m[0]
        
        fx = np.ones(n)
        for k in enumerate(m):
            fx[k[0]] = f_x.subs(x,k[1])
            
        int_num = integral(m,fx,j)
        erro = np.abs(int_num-int_f_x)
        
        if j == 1:
            mat_met1.append(int_num)
            mat_met1_e.append(erro)
            
        elif j == 2:
            mat_met2.append(int_num)
            mat_met2_e.append(erro)
        
        elif j == 3:
            mat_met3.append(int_num)
            mat_met3_e.append(erro)
        
        else:
            print('ERRO')
    mat_N.append(n)
    mat_dx.append(dx)                
```

```python
x_min = 0
x_max = 10
alpha = [0.5,1.0,1.5,2.0,2.5,3.0]

mat_N = []
mat_dx = []
mat_met_all = []
mat_met_e_all = []

for w in alpha:
    mat_N_w = []
    mat_dx_w = []
    mat_met = []
    mat_met_e = []

    x = sp.symbols('x')
    f_x = sp.exp(-w*x)*sp.sin(x)
    int_f_x = sp.integrate(f_x,(x,x1,x2))
    
    for j in range(10,120,1):
    
        m = np.linspace(x_min,x_max,j)
        f_x = sp.exp(-w*x)*sp.sin(x)
        
        fx = np.zeros(j)
        for k in enumerate(m):
            fx[k[0]] = f_x.subs(x,k[1]) 
        
        dx = m[1] - m[0]
        
        tmp = 0
        for i in range(0,j-1):
            tmp = tmp + (fx[i]+fx[i+1])*dx/2
            
        erro = np.abs(tmp-int_f_x)
        
        mat_N_w.append(j)
        mat_dx_w.append(dx)
        mat_met.append(tmp)
        mat_met_e.append(erro)
        
    mat_N.append(mat_N_w)
    mat_dx.append(mat_dx_w)
    mat_met_all.append(mat_met)
    mat_met_e_all.append(mat_met_e)
    
```

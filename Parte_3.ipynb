# Questão 1

Partindo da equação apresentada para a força magnética 

$$ \vec{F}_B = q (\vec{v} \times \vec{B}) $$

onde

$$ \vec{v} = v_x \hat{i} + v_y \hat{j} + v_z \hat{k}$$

e

$$ \vec{B} = B_x \hat{i} + B_y \hat{j} + B_z \hat{k} $$

quando realizamos o produto vetorial $(\vec{v} \times \vec{B})$ por meio do método apresentado no enunciado, obtemos um novo vetor perpendicular aos outros dois.

$$ \vec{v} \times \vec{B} = (B_zv_y - B_yv_z)\hat{i} + (B_xv_z - B_zv_x)\hat{j} + (B_yv_x - B_xv_y)\hat{k} $$

assim a força magnetica torna-se 

$$ \vec{F}_B = q\ (B_zv_y - B_yv_z)\hat{i} + q\ (B_xv_z - B_zv_x)\hat{j} + q\ (B_yv_x - B_xv_y)\hat{k} $$

Como é informado no enunciado, a força magnética é a única atuante sobre a particula, tornando-se a resultante, desse modo pela segunda lei de Newton, temos para cada um dos eixos o equilibrio de forças


$$ m(a_x \hat{i} + a_y \hat{j} + a_z \hat{k}) = q\ (B_zv_y - B_yv_z)\hat{i} + q\ (B_xv_z - B_zv_x)\hat{j} + q\ (B_yv_x - B_xv_y)\hat{k} $$

$$ m\ a_x =  q\ (B_zv_y - B_yv_z) $$

$$ m\ a_y =  q\ (B_xv_z - B_zv_x) $$

$$ m\ a_z =  q\ (B_yv_x - B_xv_y) $$

tornando $\dfrac{d\vec{v}}{dt} = \vec{a}$

$$ \dfrac{dv_x}{dt} =  \dfrac{q}{m}\ (B_zv_y - B_yv_z) $$

$$ \dfrac{dv_y}{dt} =  \dfrac{q}{m}\ (B_xv_z - B_zv_x) $$

$$ \dfrac{dv_z}{dt} =  \dfrac{q}{m}\ (B_yv_x - B_xv_y) $$

De modo a simplificar a modelagem, torna-se o campo magnético paralelo ao eixo z do sistema de referencia do movimento, desse modo $ \vec{B} = 0 \hat{i} + 0 \hat{j} + B \hat{k}$, assim teremos as demais componentes como 

$$ \dfrac{dv_x}{dt} =  \dfrac{q}{m}\ Bv_y $$

$$ \dfrac{dv_y}{dt} =  -\dfrac{q}{m}\ Bv_x $$

$$ \dfrac{dv_z}{dt} =  0 $$

esta simplificação torna a cinemática do movimento para o eixo z trivial e para os eixos x e y, um sistema de equações diferenciais acopladas. Resolvendo estas equações obtemos as equações para a velocidade para cada um dos eixos.

$$ v_x(t) =  v_{0y}\ sen(wt) + v_{0x}\ cos(wt) $$

$$ v_y(t) =  v_{0y}\ cos(wt) - v_{0x}\ sen(wt)  $$

$$ v_z(t) = v_{0z} $$

onde $ w = \dfrac{qB}{m}$ 

Tornando $\dfrac{d\vec{r}}{dt} = \vec{v}$, onde $\vec{r} = x \hat{i} + y \hat{j} + z \hat{k}$. A equação horária do movimento para cada um dos eixos será

$$ x(t) =  x_0 + \dfrac{v_{0y}}{w} + \dfrac{v_{0x}}{w}\ sen(wt) - \dfrac{v_{0y}}{w}\ cos(wt) $$

$$ y(t) =  y_0 -\dfrac{v_{0x}}{w} + \dfrac{v_{0y}}{w}\ sen(wt) + \dfrac{v_{0x}}{w}\ cos(wt) $$

$$ z(t) = z_0 + v_{0z}t $$

Resolvendo este modelo pelo método analitico por meio dos principios de Euler e Euler-Cromer na tentativo de encontrar o princípio mais apropriado para o fenômeno teremos

$$ v_x(t + \Delta t) =  v_x + \dfrac{d v_x}{dt} \Delta t =  v_x + \dfrac{q}{m}\ (B_zv_y - B_yv_z) \Delta t$$

$$ v_y(t + \Delta t) =  v_y + \dfrac{d v_y}{dt} \Delta t =  v_x + \dfrac{q}{m}\ (B_xv_z - B_zv_x) \Delta t$$

$$ v_z(t + \Delta t) =  v_z + \dfrac{d v_z}{dt} \Delta t =  v_x + \dfrac{q}{m}\ (B_yv_x - B_xv_y) \Delta t$$

pelo principio de Euler 

$$ x(t + \Delta t) =  x(t) + v_x(t) \Delta t $$

$$ y(t + \Delta t) =  y(t) + v_y(t) \Delta t $$

$$ z(t + \Delta t) =  z(t) + v_z(t) \Delta t $$

e pelo prinicipio de Euler-Cromer 

$$ x(t + \Delta t) =  x(t) + v_x(t + \Delta t) \Delta t $$

$$ y(t + \Delta t) =  y(t) + v_x(t + \Delta t) \Delta t $$

$$ z(t + \Delta t) =  z(t) + v_x(t + \Delta t) \Delta t $$

```python
# Carregando bibliotecas numérica, gráfica e simbólica

import matplotlib.pyplot as plt
import matplotlib.animation as animation
from mpl_toolkits.mplot3d import Axes3D
import scipy.integrate as spi
import numpy as np
import sympy as sp
sp.init_printing()
```

Na tentativa de determinar a melhor aproximação das equações horárias do movimento utiliza-se os métodos numéricos de Euler, de Euler-Cromer, uma variação deste ultimo e o método de Runge-kutta, definido na função odeint presente na biblioteca scipy. Desse modo, ao final da exposição dos resultados, procura-se análisar os erros de cada um dos métodos quando comparados a equações analiticas obtidas acima.

```python
def equac_analit(t0,tf,dt,vo,ro,w):
    t = np.arange(t0,tf,dt)
    n = len(t)
    vx = vo[1]*np.sin(w*t) + vo[0]*np.cos(w*t)
    vy = vo[1]*np.cos(w*t) - vo[0]*np.sin(w*t)
    vp = np.sqrt(pow(vx,2) + pow(vy,2))
    vz = [vo[2]]
    for i in range(0,n-1):   
        v_z = vo[2]
        vz.append(v_z)
        
    x = ro[0] + vo[1]/w + (vo[0]/w)*np.sin(w*t) - (vo[1]/w)*np.cos(w*t)
    y = ro[1] - vo[0]/w + (vo[1]/w)*np.sin(w*t) + (vo[0]/w)*np.cos(w*t)
    z = ro[2] + vo[2]*t
    
    return vx,vy,vz,vp,x,y,z

def metEuler(t0,tf,dt,vo,ro,B,q,m):
    t = np.arange(t0,tf,dt)
    n = len(t)
    x = [ro[0]]
    y = [ro[1]]
    z = [ro[2]]
    vx = [vo[0]]
    vy = [vo[1]]
    vz = [vo[2]]
    a = np.sqrt(pow(vo[0],2) + pow(vo[1],2))
    vp = [a]
    
    for i in range(0,n-1):
        v_x = vx[i] + ((q/m)*(B[2]*vy[i] - B[1]*vz[i]))*dt   
        v_y = vy[i] + ((q/m)*(B[0]*vz[i] - B[2]*vx[i]))*dt 
        v_z = vz[i] + ((q/m)*(B[1]*vx[i] - B[0]*vy[i]))*dt
        v_p = np.sqrt(pow(v_x,2) + pow(v_y,2))
        x_x = x[i] + vx[i]*dt
        y_y = y[i] + vy[i]*dt
        z_z = z[i] + vz[i]*dt
        
        vx.append(v_x)
        vy.append(v_y)
        vz.append(v_z)
        vp.append(v_p)
        x.append(x_x)
        y.append(y_y)
        z.append(z_z)
        
    return vx,vy,vz,vp,x,y,z

def metEulerCromer(t0,tf,dt,vo,ro,B,q,m):
    t = np.arange(t0,tf,dt)
    n = len(t)
    x = [ro[0]]
    y = [ro[1]]
    z = [ro[2]]
    vx = [vo[0]]
    vy = [vo[1]]
    vz = [vo[2]]
    a = np.sqrt(pow(vo[0],2) + pow(vo[1],2))
    vp = [a]
    
    for i in range(0,n-1):
        v_x = vx[i] + ((q/m)*(B[2]*vy[i] - B[1]*vz[i]))*dt   
        v_y = vy[i] + ((q/m)*(B[0]*vz[i] - B[2]*vx[i]))*dt 
        v_z = vz[i] + ((q/m)*(B[1]*vx[i] - B[0]*vy[i]))*dt
        v_p = np.sqrt(pow(v_x,2) + pow(v_y,2))
        x_x = x[i] + v_x*dt
        y_y = y[i] + v_y*dt
        z_z = z[i] + v_z*dt
        
        vx.append(v_x)
        vy.append(v_y)
        vz.append(v_z)
        vp.append(v_p)
        x.append(x_x)
        y.append(y_y)
        z.append(z_z)
        
    return vx,vy,vz,vp,x,y,z

def rk4(T,t):
    vx = T[0]
    vy = T[1]
    vz = T[2]
    x = T[3]
    y = T[4]
    z = T[5]
    q = 1
    m = 1
    B = [0,0,1]
    dvx = (q/m)*(B[2]*vy - B[1]*vz)
    dvy = (q/m)*(B[0]*vz - B[2]*vx)
    dvz = (q/m)*(B[1]*vx - B[0]*vy)
    dx = vx
    dy = vy
    dz = vz
    return dvx,dvy,dvz,dx,dy,dz
```

## Modelo 1

```python
t0 = 0
tf = 20
dt = 0.1
t = np.arange(t0,tf,dt)
vo = [0,1,0]     # vo = [vox,voy,voz]
ro = [1,0,0]     # ro = [xo,yo,zo]  
q = 1
m = 1
B = [0,0,1]      # B = [Bx,By,Bz]
w = q*B[2]/m     # Formulação dada quando consideramos o campo magnetico paralela ao eixo z

#Método analitico (aproximação considerando o campo magnetico constante e paralelo ao eixo z)
vx_a1,vy_a1,vz_a1,vp_a1,x_a1,y_a1,z_a1 = equac_analit(t0,tf,dt,vo,ro,w)

#Método de Euler
vx_e1,vy_e1,vz_e1,vp_e1,x_e1,y_e1,z_e1 = metEuler(t0,tf,dt,vo,ro,B,q,m)
    
#Método de Euler-Cromer
vx_ec1,vy_ec1,vz_ec1,vp_ec1,x_ec1,y_ec1,z_ec1 = metEulerCromer(t0,tf,dt,vo,ro,B,q,m)

#Método de Runge-kutta
CI = [vo[0],vo[1],vo[2],ro[0],ro[1],ro[2]]
T1 = spi.odeint(rk4,CI,t)
```

## Análise dos erros entre os modelos utilizados

```python
n = len (t)
erro_euler_x = []
erro_euler_cromer_x = []
erro_runge_kutta_x = []

erro_euler_y = []
erro_euler_cromer_y = []
erro_runge_kutta_y = []

erro_euler_z = []
erro_euler_cromer_z = []
erro_runge_kutta_z = []

for i in range(0,n):
    erro_e_x = x_a1[i] - x_e1[i]
    erro_ec_x = x_a1[i] - x_ec1[i]
    erro_rk_x = x_a1[i] - T1[:,3][i]
    
    erro_e_y = y_a1[i] - y_e1[i]
    erro_ec_y = y_a1[i] - y_ec1[i]
    erro_rk_y = y_a1[i] - T1[:,4][i]
    
    erro_e_z = z_a1[i] - z_e1[i]
    erro_ec_z = z_a1[i] - z_ec1[i]
    erro_rk_z = z_a1[i] - T1[:,5][i]
    
    erro_euler_x.append(erro_e_x)
    erro_euler_cromer_x.append(erro_ec_x)
    erro_runge_kutta_x.append(erro_rk_x)
    
    erro_euler_y.append(erro_e_y)
    erro_euler_cromer_y.append(erro_ec_y)
    erro_runge_kutta_y.append(erro_rk_y)
        
    erro_euler_z.append(erro_e_z)
    erro_euler_cromer_z.append(erro_ec_z)
    erro_runge_kutta_z.append(erro_rk_z)
```

```python
plt.figure(figsize=(20,20))

plt.subplot(3,3,1)
plt.plot(t,erro_euler_x, '-b')
plt.title("Erro Met. Euler (Eixo x)")
plt.grid()

plt.subplot(3,3,2)
plt.plot(t,erro_euler_cromer_x, '-r')
plt.title("Erro Met. Euler-Cromer (Eixo x)")
plt.grid()

plt.subplot(3,3,3)
plt.plot(t,erro_runge_kutta_x, '-g')
plt.title("Erro Met. Runge-Kutta (Eixo x)")
plt.grid()


plt.subplot(3,3,4)
plt.plot(t,erro_euler_y, '-b')
plt.title("Erro Met. Euler (Eixo y)")
plt.grid()

plt.subplot(3,3,5)
plt.plot(t,erro_euler_cromer_y, '-r')
plt.title("Erro Met. Euler-Cromer (Eixo y)")
plt.grid()

plt.subplot(3,3,6)
plt.plot(t,erro_runge_kutta_y, '-g')
plt.title("Erro Met. Runge-Kutta (Eixo y)")
plt.grid()


plt.subplot(3,3,7)
plt.plot(t,erro_euler_z, '-b')
plt.title("Erro Met. Euler (Eixo z)")
plt.grid()

plt.subplot(3,3,8)
plt.plot(t,erro_euler_cromer_z, '-r')
plt.title("Erro Met. Euler-Cromer (Eixo z)")
plt.grid()

plt.subplot(3,3,9)
plt.plot(t,erro_runge_kutta_z, '-g')
plt.title("Erro Met. Runge-Kutta (Eixo z)")
plt.grid()


plt.show()
```

Analisando os erros obtidos com a aplicação dos métodos numéricos propostos, podemos concluir que os métodos de Euler e Euler-Cromer não apresentam um bom comportamento quando dt é relativamente grande (dt>0.01), possuindo erro da primeira casa decimal, na casa de 0.5 e sempre com tendência de alta, com uma aparência de uma senoide com amplitude sempre maior. Já o metodo de Runge-Kutta apresenta um comportamento semelhante, mas com erros na ordem da sexta casa decimal.

# Questão 3

Adicionando o fator de atrito ao modelo termos

$$\vec{F}_R = \vec{F}_B - \gamma \vec{v}$$

como já aprensentado antes $\vec{v}$ continua 

$$ \vec{v} = v_x \hat{i} + v_y \hat{j} + v_z \hat{k}$$

e com resolução do produto vetorial, obtemos a força magnética $\vec{F}_B$ 

$$ \vec{F}_B = q\ (B_zv_y - B_yv_z)\hat{i} + q\ (B_xv_z - B_zv_x)\hat{j} + q\ (B_yv_x - B_xv_y)\hat{k} $$

Aplicando os principios da segunda lei de Newton a este novo arranjo obtemos

$$ m(a_x \hat{i} + a_y \hat{j} + a_z \hat{k}) = q\ (B_zv_y - B_yv_z)\hat{i} + q\ (B_xv_z - B_zv_x)\hat{j} + q\ (B_yv_x - B_xv_y)\hat{k} - \gamma v_x \hat{i} - \gamma v_y \hat{j} - \gamma v_z \hat{k} $$

o que resulta para cada eixo em 

$$ m\ a_x =  q\ (B_zv_y - B_yv_z) - \gamma v_x  $$

$$ m\ a_y =  q\ (B_xv_z - B_zv_x) - \gamma v_y $$

$$ m\ a_z =  q\ (B_yv_x - B_xv_y) - \gamma v_z $$

tornando $\vec{a_i} = \dfrac{dv_i}{dt} \hat{e_i}$

$$ \dfrac{dv_x}{dt} =  \dfrac{q}{m}\ (B_zv_y - B_yv_z) - \dfrac{\gamma}{m} v_x $$

$$ \dfrac{dv_y}{dt} =  \dfrac{q}{m}\ (B_xv_z - B_zv_x) - \dfrac{\gamma}{m} v_y $$

$$ \dfrac{dv_z}{dt} =  \dfrac{q}{m}\ (B_yv_x - B_xv_y) - \dfrac{\gamma}{m} v_z $$



```python
def rk4atrito(T,t,B,q,m,gamma):
    vx = T[0]
    vy = T[1]
    vz = T[2]
    x = T[3]
    y = T[4]
    z = T[5]
    dvx = (q/m)*(B[2]*vy - B[1]*vz) - (gamma/m)*vx
    dvy = (q/m)*(B[0]*vz - B[2]*vx) - (gamma/m)*vy
    dvz = (q/m)*(B[1]*vx - B[0]*vy) - (gamma/m)*vz
    dx = vx
    dy = vy
    dz = vz
    return dvx,dvy,dvz,dx,dy,dz
```
